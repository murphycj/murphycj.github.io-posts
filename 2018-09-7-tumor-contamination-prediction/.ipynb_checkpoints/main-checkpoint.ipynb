```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

```python
counts = pd.read_csv('./data/BedtoolsCount.counts.csv')
counts_trp53 = pd.read_csv('./data/BedtoolsCount.Trp53.counts.csv')
```

```python
counts.columns = counts.columns.str.replace('-','')
```

```python

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(10,5)

ax1.scatter(counts['HLWES41'],counts['HLWES18'],s=1)
ax1.set_xlim(0,25000) #outliers distort view
ax1.set_ylim(0,25000)
ax1.set_xlabel('Liver sample 1 read counts')
ax1.set_ylabel('Tumor sample 1 read counts')
ax1.set_title('Counts')

ax2.scatter(np.log2(counts['HL-WES-40']+1),np.log2(counts['HL-WES-18']+1),s=1)
ax2.set_xlabel('log2(Liver sample 1 read counts)')
ax2.set_ylabel('log2(Tumor sample 1 read counts)')
ax2.set_title('log(Counts+1)')

plt.tight_layout()
plt.savefig('./plots/scatter.png')
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from patsy import dmatrices
```

```python
y, X = dmatrices('HLWES18 ~ HLWES40', data=counts, return_type='dataframe')
```

```python
poisson_model = sm.GLM(y, X, family=sm.families.Poisson())
poisson_model_results = poisson_model.fit()
print(poisson_model_results.summary())
```

```python
fig = poisson_model_results.plot_partial_residuals('HLWES40')
plt.xlim(0,20000)
```

```python
y, X = dmatrices('HLWES18 ~ log(HLWES40+1)', data=counts, return_type='dataframe')
poisson_model = sm.GLM(y, X, family=sm.families.Gassian())
poisson_model_results = poisson_model.fit()
print(poisson_model_results.summary())
```

```python
poisson_model_results.summary2()
```

```python
plt.scatter(X['HLWES40'],poisson_model_results.resid_deviance,s=1)
plt.xlim(0,20000)
```
